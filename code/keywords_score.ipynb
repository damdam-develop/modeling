```python
from notebook_model import test_set, tokenize, term_frequency

import re
import pandas as pd
import numpy as np
from konlpy.tag import Okt

okt = Okt()

test_set
```

```python
# 동의어 사전
# 글자수가 같은 명사만 들어가야 함
keywords_list = {'가성':['가격', '무난'], # 가성비를 토큰화하면 가성+비로 나눠짐
                'AS':['수리', '서비'],
                '사은':['추가'],  # 사은품을 토큰화하면 사은+품으로 나눠짐
                '배송':['기사', '발송'],
                '성능':['속도', '힌지', '포트', '사용']}
```

```python
cleanzing = []
for doc in test_set['content']:
    # 영어 소문자를 대문자로 통일
    doc = doc.upper()
    # 한글 자음, 모음 제거
    doc = re.sub(pattern='([ㄱ-ㅎㅏ-ㅣ]+)', repl=' ', string=doc)
    cleanzing.append(doc)

df = pd.DataFrame(cleanzing, columns = ['reviews'])
df
```

```python
for keyword in keywords_list:
    df[keyword] = -1.0
df
```

```python
def index_token(review):
    s_copy = review

    token = okt.pos(s_copy)

    # 토큰화 후 키워드 유의어 사전을 이용해 키워드로 바꿔준다
    # 그리고 인덱싱도 찾아준다.
    # index는 s에서 토큰의 위치
    index = 0
    new_token = []
    for i, j in enumerate(token):
        new_token.append([j[0], j[1], index])
        for keyword, related_words in keywords_list.items():
            if j[0] in related_words:
                new_token[i][0] = keyword

        # index에 현재 토큰의 길이를 더해주고 s_copy에서 토큰을 없애준다.
        index += len(j[0])
        s_copy = s_copy.replace(j[0],'', 1)

        # 공백이 있으면 공백의 개수만큼 index에 더해주고 공백을 없애준다.
        if len(s_copy)>0 and s_copy[0] == ' ':
            index += 1
            s_copy = s_copy.replace(' ','', 1)
    return new_token
```

```python
def delete_token(tenwords, poomsa):
    for j in range(1, len(tenwords)):
        if tenwords[j][1] == poomsa:
            return 1, tenwords[:j+1]
    return 0, tenwords
```

```python
from keras.models import load_model

model = load_model('notebook_model.h5')

def predict_score(review):
    # 키워드가 없을 시 score는 -1
    if not review:
        score = -1
    else:
        token = tokenize(review)
        tf = term_frequency(token)
        data = np.expand_dims(np.asarray(tf).astype('float32'), axis=0)
        score = float(model.predict(data))
    return score
```

```python
for row, review in enumerate(df['reviews']):
    new_token = index_token(review)
    for k in keywords_list.keys():
        index = 0
        highlight = []
        for i, j in enumerate(new_token):
            if(j[0] == k):    # 특정 키워드가 있는지 확인
                if(i>0 and new_token[i-1][1] == 'Adjective'): # 키워드의 앞 단어가 형용사이면 highlight에 추가
                    highlight.append(new_token[i-1][0])

                # 키워드부터 10 단어 자르기
                tenwords = new_token[i:i+10]

                # tenwords에서 품사가 기호, 형용사, 동사인 순서로 자르기
                for poomsa in ["Punctuation", "Adjective", "Verb"] :
                    zero_one, tmp_tenwords = delete_token(tenwords, poomsa)
                    # 위에 중 하나라도 만났으면 for문 나가기
                    if zero_one == 1 :
                        break;

                # tmp_tenwords에서 시작과 끝 단어를 리뷰에서 찾아서 highlight로 뽑아낸다.
                start = tmp_tenwords[0][2]
                end = tmp_tenwords[-1][2] + len(tmp_tenwords[-1][0])
                highlight.append(review[start:end])
                index += 1

        my_string = ', '.join(highlight)
        score = predict_score(my_string)
        df[k][row] = round(score,4)
```
